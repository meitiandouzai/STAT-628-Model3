```python
import os
import pandas as pd
from pytz import timezone
from datetime import datetime
import pytz
```

```python
# set match: AIRPORT & Timezone
timezones_df = pd.read_csv('airport_timezones2.csv', usecols=['AIRPORT', 'Timezone'])
match = dict(zip(timezones_df['AIRPORT'], timezones_df['Timezone']))

# target timezone
target_timezone = timezone('America/Chicago')

# TODO: change file folder
folder_path = r'D:\lessons\stat628\project3_flight\climate2024'
target_path = r'D:\lessons\stat628\project3_flight\changetime_climate2024'
os.makedirs(target_path, exist_ok=True)
```

```python
#修改LCD_OAK_2019
file = r'D:\lessons\stat628\project3_flight\LCD_LYH_2019.csv'
try:
    chunks = pd.read_csv(file, chunksize=1000, on_bad_lines='skip')
    climate_data = pd.concat(chunks, ignore_index=True)

    # 保留11月到1月的记录
    climate_data['DATE'] = pd.to_datetime(climate_data['DATE'], errors='coerce')
    climate_data = climate_data[climate_data['DATE'].dt.month.isin([11, 12, 1])]
    
    # 保存处理后的文件
    climate_data.to_csv(file, index=False)
    print(f"Processed file: {file}")

except Exception as e:
    print(f"Error processing file {file}: {e}")
```

```python
expected_format = '%Y/%m/%d %H:%M:%S'

for file_name in os.listdir(folder_path):
    if file_name.startswith('LCD_') and file_name.endswith('.csv'):
        # extract data in folder
        airport_code = file_name.split('_')[1]

        # check if data has already changed
        output_file_name = file_name
        if os.path.exists(os.path.join(target_path, output_file_name)):
            print(f"文件已存在，跳过: {output_file_name}")
            continue

        # obtain original timezone
        if airport_code in match:
            source_timezone = timezone(match[airport_code])

            # read data
            lcd_df = pd.read_csv(os.path.join(folder_path, file_name))
            converted_dates = []

            # check date format
            for date_str in lcd_df['DATE']:
                try:
                    dt = datetime.strptime(date_str, expected_format)
                    localized_dt = source_timezone.localize(dt).astimezone(target_timezone)
                    converted_dates.append(localized_dt.strftime(expected_format))
                except ValueError:
                    converted_dates.append(date_str)

            lcd_df['DATE'] = converted_dates

            # save data
            lcd_df.to_csv(os.path.join(target_path, output_file_name), index=False)
            print(f"转换完成: {output_file_name}")
```
